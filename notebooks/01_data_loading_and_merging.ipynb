## Loading the accident data and merging
 them


```python
import pandas as pd

years = [2020,2021,2022,2023]
dfs = []

for year in years:
    print(year)
    try:
        df = pd.read_csv(f"../data/{year}/accident.csv",encoding="cp1252",low_memory=False)
        df['YEAR'] = year
        dfs.append(df)
    except FileNotFoundError:
        print(f"{year}: file not found")
accident_dfs = pd.concat(dfs,axis = 0,ignore_index=True)
```

```python
# Finalized data_frame 
accident_dfs.head()

# Info about accident_dfs

print(f"Total rows: {len(accident_dfs)}")
print(f"Columns: {len(accident_dfs.columns.tolist())}")
print(f"Years: {accident_dfs['YEAR'].value_counts().sort_index()}")
print(f"WEATHERNAME missing: {accident_dfs['WEATHERNAME'].isna().sum()}")
```
