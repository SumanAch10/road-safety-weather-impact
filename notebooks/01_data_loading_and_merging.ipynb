## Loading the accident data and merging
 them


```python
import pandas as pd

years = [2020,2021,2022,2023]
dfs = []

for year in years:
    try:
        df = pd.read_csv(f"../data/{year}/accident.csv",encoding="cp1252",low_memory=False)
        df['YEAR'] = year
        dfs.append(df)
    except FileNotFoundError:
        print(f"{year}: file not found")
accident_dfs = pd.concat(dfs,axis = 0,ignore_index=True)

# print(accident_dfs.describe())
```

## Loading the weather data and merging them

```python
years = [2020,2021,2022,2023]
dfs = []

for year in years:
    try:
        df = pd.read_csv(f"../data/{year}/weather.csv")
        df['YEAR'] = year
        dfs.append(df)
    except FileNotFoundError:
        print(f"{year}: No weather data found")
weather_dfs = pd.concat(dfs,ignore_index=True)
print(weather_dfs.groupby('YEAR')['WEATHERNAME'].value_counts())
```
